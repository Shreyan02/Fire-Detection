```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import random
import csv
```

```python
dataset = pd.read_csv("sdcReduced3.csv")
```

```python
dataset
```

```python
# input
x = dataset.iloc[:, 0:4].values

# output
y = dataset.iloc[:, 5].values
```

```python
# importing utility modules
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss

# importing machine learning models for prediction
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

from sklearn import model_selection
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn. preprocessing import StandardScaler

# importing voting classifier
from sklearn.ensemble import VotingClassifier
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.30, random_state=0
)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
model1 = GaussianNB()
model2 = LogisticRegression()
model3 = RandomForestClassifier()
model4 = DecisionTreeClassifier()
model5 = KNeighborsClassifier(n_neighbors=2)
model6 = SVC()
```

```python
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)
model4.fit(X_train, y_train)
model5.fit(X_train, y_train)
model6.fit(X_train, y_train)
```

```python
y_pred1 = model1.predict(X_test)
y_pred2 = model2.predict(X_test)
y_pred3 = model3.predict(X_test)
y_pred4 = model4.predict(X_test)
y_pred5 = model5.predict(X_test)
y_pred6 = model6.predict(X_test)
```

```python
labels = ["0", "1"]
cm_NaiveBayes = confusion_matrix(y_test, y_pred1)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_NaiveBayes, display_labels=labels)
disp.plot();
```

```python
labels = ["0", "1"]
cm_LogisticRegression = confusion_matrix(y_test, y_pred2)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_LogisticRegression, display_labels=labels)
disp.plot();
```

```python
labels = ["0", "1"]
cm_RandomForest = confusion_matrix(y_test, y_pred3)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_RandomForest, display_labels=labels)
disp.plot();
```

```python
labels = ["0", "1"]
cm_DecisionTree = confusion_matrix(y_test, y_pred4)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_DecisionTree, display_labels=labels)
disp.plot();
```

```python
labels = ["0", "1"]
cm_KNN = confusion_matrix(y_test, y_pred5)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_KNN, display_labels=labels)
disp.plot();
```

```python
labels = ["0", "1"]
cm_SupportVectorClass = confusion_matrix(y_test, y_pred6)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_SupportVectorClass, display_labels=labels)
disp.plot();
```

```python
result1 = model_selection.cross_val_score(model1, X_train, y_train)
result2 = model_selection.cross_val_score(model2, X_train, y_train)
result3 = model_selection.cross_val_score(model3, X_train, y_train)
result4 = model_selection.cross_val_score(model4, X_train, y_train)
result5 = model_selection.cross_val_score(model5, X_train, y_train)
result6 = model_selection.cross_val_score(model6, X_train, y_train)
```

```python
print('Accuracy of Naive Bayes Model = ',result1.mean())
print('Accuracy of Logistic Regression Model = ',result2.mean())
print('Accuracy of Random Forest Model = ',result3.mean())
print('Accuracy of Decision Tree Model = ',result4.mean())
print('Accuracy of k-NN Model = ',result5.mean())
print('Accuracy of Support Vector Model = ',result6.mean())
```

```python
from sklearn.metrics import accuracy_score

print('Accuracy of Naive Bayes Model = ',accuracy_score(y_test, y_pred1))
print('Accuracy of Logistic Regression Model = ',accuracy_score(y_test, y_pred2))
print('Accuracy of Random Forest Model = ',accuracy_score(y_test, y_pred3))
print('Accuracy of Decision Tree Model = ',accuracy_score(y_test, y_pred4))
print('Accuracy of k-NN Model = ',accuracy_score(y_test, y_pred5))
print('Accuracy of Support Vector Model = ',accuracy_score(y_test, y_pred6))
```

```python
print(log_loss(y_test, y_pred1))
print(log_loss(y_test, y_pred2))
print(log_loss(y_test, y_pred3))
print(log_loss(y_test, y_pred4))
print(log_loss(y_test, y_pred5))
print(log_loss(y_test, y_pred6))
```

```python
final_model = VotingClassifier(
	estimators=[('rf', model3), ('dt', model4)], voting='hard')
final_model.fit(X_train, y_train)
```

```python
pred_final = final_model.predict(X_test)
```

```python
labels = ["0", "1"]
cm_final = confusion_matrix(y_test, pred_final)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_final, display_labels=labels)
disp.plot();
```

```python
resultfinal = model_selection.cross_val_score(final_model, X_train, y_train)
print('Accuracy of Ensemble Model = ',resultfinal.mean())
```

```python
from sklearn.metrics import accuracy_score

print ("Accuracy : ", accuracy_score(y_test, pred_final))
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred1, average='macro')
print("F1 Score NB:", f1)
f2 = f1_score(y_test, y_pred2, average='macro')
print("F1 Score LR:", f2)
f3 = f1_score(y_test, y_pred3, average='macro')
print("F1 Score RF:", f3)
f4 = f1_score(y_test, y_pred4, average='macro')
print("F1 Score DT:", f4)
f5 = f1_score(y_test, y_pred5, average='macro')
print("F1 Score KNN:", f5)
f6 = f1_score(y_test, y_pred6, average='macro')
print("F1 Score SVC:", f6)
ffinal = f1_score(y_test, pred_final, average='macro')
print("F1 Score:", ffinal)
```
